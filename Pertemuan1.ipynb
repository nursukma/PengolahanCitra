```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #digunakan untuk menampilkan gambar
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

#ini adalah link imagenya
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]

for url in urls:
  image = io.imread(url) #ini digunakan untuk read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #ini digunakan untuk mengatur ukuran
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #ini untuk convert warna

  tinggi = image_2.shape[0] #shape.0 merupakan tinggi
  lebar = image_2.shape[1] #shape.1 merupakan lebar
  print('resolusi: tinggi x lebar -> ', tinggi, lebar)

  #ini digunakan untuk membuat garis horizontal di tengah image
  image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #ini untuk convert warna
  for y in range (lebar):
    image_3[int((tinggi)/2),y] = [255,255,255]

  final_frame = cv.hconcat((image ,image_2, image_3)) #ini adalah untuk membuat frame
  cv2_imshow(final_frame) #print image
  print('\n')
```
